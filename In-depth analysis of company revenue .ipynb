# We done with all cleaning  and now data is clean and organized in data format. Now we can go for the data analysis and visualization

```python
g= df.groupby(['Country'],as_index = False)['Gross Sales'].sum().sort_values(by='Gross Sales',ascending=False)
g
```

```python
group = df.groupby('Country').agg({'Sales':'sum','Profit':'sum','Units Sold':'sum','COGS':'sum'}).reset_index()
```

```python
group
```

```python
sns.set(rc={'figure.figsize':(10 , 5)})
sns.barplot(x = "Country",y= "Sales", data= group,label= "Sales", color='#BEADFA')
plt.show()
```

#We can see that highest sale is noticed in USA and in Canada and lowest sale is recorded in Mexico


```python
plt.figure(figsize=(10,6))
bar_width=0.35
x= np.arange(len(group['Country']))
plt.bar(x-bar_width/2,group['Sales'],bar_width,label='Sales',color='#BEADFA')
plt.bar(x + bar_width/2,group['COGS'],bar_width, label='COGS',color='#FFC0D9')
plt.xlabel('Country')
plt.ylabel('Values')
plt.xticks(x,group['Country'])
plt.legend()
plt.show()
```

# We can see that Sales is more than the COGS(Cost of Gross Sale) it means company is in profit


```python
plt.figure(figsize=(10,6))
c= ('#AC87C5','#FF90BC','#F6F7C4','#A1EEBD','#7BD3EA')
plt.pie(group['Profit'],labels = group['Country'],autopct='%1.1f%%',startangle=140, colors= c)
plt.axis('equal')
plt.title('Profit Distribution by Country')
plt.show()
```


# Product wise analysis

```python
group_product = df.groupby('Product').agg({
    'Units Sold': 'sum',
    'Manufacturing Price': 'mean',
    'Sale Price': 'mean',
    'Gross Sales': 'sum',
    'Profit': 'sum',
    'COGS': 'sum'
}).reset_index()
```

```python
# we can create one more column which contain margin that is selling price minus manufacturing price
group_product['Gross Margin']=group_product['Sale Price']- group_product['Manufacturing Price']
```

```python
categories=group_product['Product']
margin_values=group_product['Gross Margin']
```

```python
group_product
```

```python
categories=group_product['Product']
margin_values=group_product['Gross Margin']
sns.barplot(x=categories, y=margin_values, label='Gross Margin',color='#BEADFA')

plt.show()
```

#From above figure we can see that Product 'Amarilla' and product 'VTT' has negative gross margin and product 'Velo' has very low margin

```python
c= ('#AC87C5','#FF90BC','#F6F7C4','#A1EEBD','#7BD3EA','#FFC5C5')
plt.pie(group_product['Profit'],labels=group_product['Product'],autopct='%1.1f%%',startangle= 140,colors=c)
plt.title('Product wise Profit')
plt.show()
```

# From here we can see that Paseo has highest profit and Montana have low profit

```python

# Reshape the DataFrame to long form
df_long = pd.melt(group_product, id_vars=['Product'], value_vars=['Gross Sales', 'COGS'], 
                  var_name='Category', value_name='Values')
category_colors = {'Gross Sales': '#BEADFA', 'COGS': '#FFC0D9'}

plt.figure(figsize=(10, 6))
sns.barplot(data=df_long, x='Product', y='Values', hue='Category',palette=category_colors)
plt.xlabel('Product')
plt.ylabel('Values')
plt.title('Sales and COGS per Product')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()
```

```python

# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' column as index
df.set_index('Date', inplace=True)

# Now, 'Date' is the index with DateTimeIndex

# Resample the sales data to get monthly and quarterly sales
monthly_sales = df['Sales'].resample('M').sum()
quarterly_sales = df['Sales'].resample('Q').sum()

# Create the plot
plt.figure(figsize=(14, 5))

# Monthly Line Graph
plt.subplot(1, 2, 1)
plt.plot(monthly_sales, marker='o',color= '#EA8FEA')
plt.title('Monthly Sales')

# Quarterly Line Graph
plt.subplot(1, 2, 2)
plt.plot(quarterly_sales, marker='o',color= '#EA8FEA')
plt.title('Quarterly Sales')

# Show the plots
plt.tight_layout()
plt.show()
```

```python

monthly_profit = df['Profit'].resample('M').sum()
quarterly_profit = df['Profit'].resample('Q').sum()

# Create line graphs for monthly and quarterly sales
plt.figure(figsize=(14, 5))

# Monthly line graph
plt.subplot(1, 2, 1)
plt.plot(monthly_profit, marker='o',color= '#EA8FEA')
plt.title('Monthly Profit')

# Quarterly line graph
plt.subplot(1, 2, 2)
plt.plot(quarterly_profit, marker='o',color= '#EA8FEA')
plt.title('Quarterly Profit')

# Show the plots
plt.tight_layout()
plt.show()
```
